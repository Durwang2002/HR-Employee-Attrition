#Feature Engineering

```python
#Create Age Group
```

```python
age_bins = [0, 25, 35, 45, 55, 65, 100]
age_labels = ["18-25", "26-35", "36-45", "46-55", "56-65", "65+"]
hr["AgeGroup"] = pd.cut(hr["Age"], bins=age_bins, labels=age_labels)
hr["AgeGroup"].value_counts()
```

```python
#Monthly Income Category
```

```python
income_bins = [0, 3000, 7000, 12000, 20000]

income_labels = [
    "Low",
    "Medium",
    "High",
    "Very High"
]

hr["IncomeCategory"] = pd.cut(
    hr["MonthlyIncome"],
    bins=income_bins,
    labels=income_labels
)

hr["IncomeCategory"].value_counts()
```

```python
#Experience Group
```

```python
experience_bins = [-1, 5, 10, 20, 40]

experience_labels = [
    "0-5 Years",
    "6-10 Years",
    "11-20 Years",
    "20+ Years"
]

hr["ExperienceGroup"] = pd.cut(
    hr["TotalWorkingYears"],
    bins=experience_bins,
    labels=experience_labels
)

hr["ExperienceGroup"].value_counts()
```

```python
#Tenure Group
```

```python
tenure_bins = [-1, 2, 5, 10, 20, 50]

tenure_labels = [
    "0-2 Years",
    "3-5 Years",
    "6-10 Years",
    "11-20 Years",
    "20+ Years"
]

hr["TenureGroup"] = pd.cut(
    hr["YearsAtCompany"],
    bins=tenure_bins,
    labels=tenure_labels
)

hr["TenureGroup"].value_counts()
```

```python
#Promotion Status
```

```python
hr["PromotionStatus"] = np.where(
    hr["YearsSinceLastPromotion"] >= 5,
    "Long Time Without Promotion",
    "Recently Promoted"
)

hr["PromotionStatus"].value_counts()
```

```python
#Overtime Flag
```

```python
hr["OverTimeFlag"] = hr["OverTime"].map({
    "Yes": 1,
    "No": 0
})

hr["OverTimeFlag"].value_counts()
```

```python
#Income Per Working Year
```

```python
hr["IncomePerYearExperience"] = round(hr["MonthlyIncome"]/(hr["TotalWorkingYears"] + 1),2)

hr[
    [
        "MonthlyIncome",
        "TotalWorkingYears",
        "IncomePerYearExperience"
    ]
].head()
```

```python
#Attrition Flag
```

```python
hr["AttritionFlag"] = hr["Attrition"].map({
    "Yes": 1,
    "No": 0
})

hr["AttritionFlag"].value_counts()
```

```python
#Verify New Features
```

```python
new_features = [

    "AgeGroup",

    "IncomeCategory",

    "ExperienceGroup",

    "TenureGroup",

    "PromotionStatus",

    "OverTimeFlag",

    "IncomePerYearExperience",

    "AttritionFlag"

]

hr[new_features].head()
```

```python
#Final Dataset Shape
```

```python
print("Final Shape :", hr.shape)
```

```python
#Feature Engineering Summary
```

```python
feature_summary = pd.DataFrame({

    "Feature Created":[

        "AgeGroup",

        "IncomeCategory",

        "ExperienceGroup",

        "TenureGroup",

        "PromotionStatus",

        "OverTimeFlag",

        "IncomePerYearExperience",

        "AttritionFlag"

    ],

    "Purpose":[

        "Employee Segmentation",

        "Salary Segmentation",

        "Experience Analysis",

        "Retention Analysis",

        "Career Growth Analysis",

        "Binary Analysis",

        "Income Efficiency",

        "Target Encoding"

    ]

})

feature_summary
```

Feature Engineering Insights
1. Age Group

Employees were segmented into age groups to identify which age category has the highest attrition rate.

2. Income Category

Monthly income was categorized into salary bands, making it easier to compare attrition across income levels.

3. Experience Group

Employees were grouped by total working experience to analyze whether early-career or experienced employees are more likely to leave.

4. Tenure Group

Company tenure was categorized to study employee retention at different stages of employment.

5. Promotion Status

Employees who had not received a promotion for five or more years were identified to evaluate the impact of career growth on attrition.

6. Overtime Flag

The overtime column was converted into a binary feature to simplify statistical analysis and correlation.

7. Income per Year of Experience

A derived metric was created to compare salary relative to overall work experience.

8. Attrition Flag

The target variable was encoded into binary format (Yes = 1, No = 0) for correlation analysis, hypothesis testing, and machine learning compatibility.

💼 Business Impact
Employee segmentation enables targeted HR policies.
Salary bands help evaluate compensation fairness.
Experience and tenure groups identify high-risk retention segments.
Promotion analysis supports career planning initiatives.
Binary features simplify advanced statistical analysis and predictive modeling.

#Statistical Testing

```python
#Import Statistical Libraries
```

```python
from scipy.stats import ttest_ind
from scipy.stats import chi2_contingency
```

```python
#Independent T-Test
```

```python
stay = hr.loc[
    hr["Attrition"] == "No",
    "MonthlyIncome"
]

leave = hr.loc[
    hr["Attrition"] == "Yes",
    "MonthlyIncome"
]

t_stat, p_value = ttest_ind(
    stay,
    leave,
    equal_var=False
)

print("T-Statistic :", round(t_stat,3))
print("P-Value :", round(p_value,5))
```

```python
alpha = 0.05

if p_value < alpha:
    print("Reject Null Hypothesis")
    print("Monthly Income significantly differs between employees who stayed and those who left.")
else:
    print("Fail to Reject Null Hypothesis")
    print("No statistically significant difference found.")
```

```python
#Business Insight
#If p < 0.05, salary is statistically associated with employee attrition.
#HR should review compensation policies, especially for high-turnover employee groups.
```

```python
#Chi-Square Test
```

```python
contingency_table = pd.crosstab(
    hr["OverTime"],
    hr["Attrition"]
)

contingency_table
```

```python
chi2, p_value, dof, expected = chi2_contingency(
    contingency_table
)

print("Chi-Square :", round(chi2,3))
print("P-Value :", round(p_value,5))
```

```python
alpha = 0.05

if p_value < alpha:
    print("Reject Null Hypothesis")
    print("Overtime and Attrition are statistically associated.")
else:
    print("Fail to Reject Null Hypothesis")
    print("No statistical association found.")
```

```python
#Insight
#If p < 0.05, overtime has a statistically significant relationship with employee attrition.
#HR should evaluate overtime policies and workload distribution.
```

#Anomaly Detection

```python
#Detect Outliers using IQR
```

```python
def detect_outliers(data, column):

    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = data[
        (data[column] < lower) |
        (data[column] > upper)
    ]

    return outliers

income_outliers = detect_outliers(hr, "MonthlyIncome")

print("Monthly Income Outliers :", len(income_outliers))

income_outliers.head()
```

#Confidence Intervals

```python
#Import Required Library
from scipy.stats import t
```

```python
#Function to Calculate 95% Confidence Interval
```

```python
def confidence_interval(series, confidence=0.95):

    mean = series.mean()
    std = series.std()
    n = len(series)

    margin = t.ppf(
        (1 + confidence) / 2,
        df=n - 1
    ) * (std / np.sqrt(n))

    lower = mean - margin
    upper = mean + margin

    return round(mean,2), round(lower,2), round(upper,2)
```

```python
#Monthly Income Confidence Interval
```

```python
mean, lower, upper = confidence_interval(
    hr["MonthlyIncome"]
)

print(f"Mean Monthly Income : {mean}")
print(f"95% Confidence Interval : ({lower}, {upper})")
```

```python
#Age Confidence Interval
```

```python
mean, lower, upper = confidence_interval(
    hr["Age"]
)

print(f"Mean Age : {mean}")
print(f"95% Confidence Interval : ({lower}, {upper})")
```

```python
#Job Satisfaction Confidence Interval
```

```python
mean, lower, upper = confidence_interval(
    hr["JobSatisfaction"]
)

print(f"Mean Job Satisfaction : {mean}")
print(f"95% Confidence Interval : ({lower}, {upper})")
```
